# Accessing items from a list

```python
#indexing
L=[1,2,3,4,5,6]
A=[1,2,3,[4,5]]
B=[[[1,2],[3,4]],[[5,6],[7,8]]]

#positve
print(L[0])
print(L[3])
print(A[3][1])
print(B[1])
print(B[1][0][0])
print(B[0][0][1])

#negative
print(L[-1])
print(L[-3])
print(A[-1])
print(A[-1][-2])
print(B[-1][-2][-1])
print(B[-2][-1][-1])


#slicing
print(L[0:3])
print(L[-3:])
print(L[0::2])
print(L[-5:-2:2])
```

# Adding items to a List

```python
#append->add at end
L=[1,2,3,4,5,6]
L.append(True)
print(L)


L.append([6,7,8])
print(L)

L.append('pune')
print(L)
```

```python
#extend->add multiple items
L=[1,2,3,4,5,6]
L.extend([6,7,8])
print(L)

L.extend('pune')
print(L)
```

```python
#insert
L=[1,2,3,4,5,6]
L.insert(1,100)
print(L)
```

# Editing items in list

```python
L=[1,2,3,4,5]

L[1:4]=[200,300,400]
print(L)

L[-1]=500
print(L)
```

# Deleting items from a list

```python
# del
L=[1,2,3,4,5]
print(L)

del L[-1]
print(L)
del L[1:3]
print(L)
del L
print(L)
```

```python
# remove
L=[1,2,3,4,5]
print(L)

L.remove(5)
print(L)
```

```python
# pop
L=[1,2,3,4,5]
print(L)

L.pop(0)
print(L)

L.pop() #delete last element
print(L)
```

```python
#clear
L=[1,2,3,4,5]
print(L)

L.clear()
print(L)
```

# operations on Lists

```python
#arithmetic(+,*)
L1=[1,2,3,4]
L2=[5,6,7]
#CONCATINATE/MERGE
print(L1+L2)

print(L1*3)
```

```python
#membership
L1=[1,2,3,4]
L2=[1,2,3,4,[5,6]]

print(5 in L1)
print(5 in L2)
print([5,6] in L2)
print(5 not in L1)
print(7 not in L2)
```

# List Function

```python
#len/min/max/sorted
L1=[1,2,3,4]
print(len(L1))
print(min(L1))
print(max(L1))
print(sorted(L1))
print(sorted(L1,reverse=True))
```

```python
#count
L1=[1,2,3,4,1,2,4,3,4,5,5,6,6,7]
print(L1.count(4))
print(L1.count(7))
print(L1.count(2))
```

```python
#index
L1=[1,2,3,4,1,4]
print(L1.index(3))
print(L1.index(1))
```

```python
#reverse
L1=[1,2,3,4,1,4]
print(L1.index(3))
print(L1.index(1))

#permanent operation->reverse
L1.reverse()
print(L1)
```

```python
#sort(vs sorted)
L1=[1,2,3,4,1,4]
print(L1)
print(sorted(L1))  #temperoy
print(L1)
L1.sort()  #permenent
print(L1)
```

```python
#copy-<shallow
L=[1,2,3,4,0]
print(L)
print(id(L))
L1=L.copy()
print(L1)
print(id(L1))
```

# List Comprehension


List Comprehension


List Comprehension provides a concise way of creating lists.



newlist = [expression for item in iterable if condition == True]image.png



Advantages of List Comprehension



More time-efficient and space-efficient than loops.


Require fewer lines of code.


Transforms iterative statement into a formula.

```python
#add 1 to 10 number to a list
L=[]
for i in range(1,11):
    L.append(i)
print(L)
```

```python
L=[i for i in range(1,11)]
print(L)
```

```python
#scaler multiplication on a vector
v=[2,3,4]
s=-3
#[-6,-9,-12]
x=[]
for i in v:
    x.append(i*s)
print(x)
```

```python
[s*i for i in v]
```

```python
# add square
L=[1,2,3,4,5]
[i**2 for i in L]
```

```python
#print all numbers divisible by 5 in the range of 1 to 50
[i for i in range(1,50) if i%5==0]
```

```python
#find language which start with letter p
language=['java','c++','python','c','php','html']
[language for language in language if language[0]=='p']
#or
[language for language in language if language.startswith('p')]
```

```python
#nested if with list Comprehension
basket=['apple','guava','cherry','banana']
my_fruits=['apple','kiwi','graps','banana']

#add new list from my_fruits and items if the fruit exist in basket and also start with 'a'
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')] 
```

```python
#print a(3,3) matrix using list comprehension ->nested list comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian products-> list comprehension on 2 lits together
L1=[1,2,3,4]
L2=[5,6,7,8]

[i*j for i in L1 for j in L2]
```
